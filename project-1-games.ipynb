<p>Проанализируем типы данных

<p>В колонках <code>Year of Release</code>, <code>NA sales</code>, <code>Other sales</code>, <code>Critic Score</code> к типу данных вопросов нет. А вот в остальных интересно проверить, что там находиться по типу данных, поскольку тип <code>object</code> может говорить о нескольких типах данных в столбце.</p>
<p>Поэтому, проверим, что там находится.</p>

```python
object_types_columns = ['name', 'year_of_release', 'platform', 'genre', 'eu_sales', 'jp_sales', 'user_score', 'rating']

def check_unique_column_types(column_name):
    unique_types = df[f'{column_name}'].apply(type).unique().tolist()
    print(f'Используемые типы данных в столбце {column_name} - {unique_types}')

for element in object_types_columns:
    check_unique_column_types(element)
```

Теперь поработаем с присвоением нужного типа данных, учитывая комментарий ниже, что могу встречаться строковые значения в числовых столбцах. Также стоит не забыть наши изменения выше, т.е. нам нужно теперь только поработать со столбцами <code>eu_sales</code>, <code>jp_sales</code>, <code>user_score</code> и <code>rating</code>.

Перед переводом к нужным типам интересно посмотреть, что за данные вообще могут быть в столбце, отличные от float или object

```python
columns_to_float = ['eu_sales', 'jp_sales', 'user_score']

def check_float_type(column_name):
    float_values = set()
    for e in df[column_name]:
        if isinstance(e, float):
            float_values.add(e)
    print(f'Значения с типом float в {column_name} - {float_values}')

def check_numeric(column_name):
    non_numeric = set()
    for e in df[column_name]:
        try:
            float(e)
        except ValueError:
            non_numeric.add(e)
    print(f'Не конвертируемые значения {column_name} - {non_numeric}')  

for element in columns_to_float:
    check_numeric(element)
    
check_float_type('rating')
```

Видим, что ничего критичного в "ненужных" типах данных нет и можно смело менять типы данных в столбцах

Переведем столбцы `eu_sales`,`jp_sales` и `user_score` к типу `float`

```python
df[['eu_sales', 'jp_sales', 'user_score']] = df[['eu_sales', 'jp_sales', 'user_score']].apply(pd.to_numeric, errors='coerce')
```

Попробуем посчитать относительную разницу между средним и медианой

## Категоризация данных

Разделим все игры по оценкам пользователей и выделим такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
def categorize_game_user_score(row):
    if row['user_score'] >= 8:
        return "Высокая оценка"
    elif row['user_score'] >= 3:
        return "Средняя оценка"
    else:
        return "Низкая оценка"

df_actual['user_category'] = df_actual.apply(categorize_game_user_score, axis=1)
df_actual
```

Разделим все игры по оценкам критиков и выделим такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
def categorize_game_critic_score(row):
    if row['critic_score'] >= 80:
        return "Высокая оценка"
    elif row['critic_score'] >= 30:
        return "Средняя оценка"
    else:
        return "Низкая оценка"

df_actual['critic_category'] = df_actual.apply(categorize_game_critic_score, axis=1)
df_actual
```

Проверим результат - сгруппируйем данные по выделенным категориям и посчитаем количество игр в каждой категории.

```python
grouped_user_category = df_actual.groupby('user_category')['name'].count()
grouped_user_category
```

```python
grouped_critic_category = df_actual.groupby('critic_category')['name'].count()
grouped_critic_category
```
